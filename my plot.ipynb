```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Adepoju Wasiu\Downloads\customers.csv')
df.head(10)
```

```python
df.isna().sum()
df.shape
high = df[df['Spending Score (1-100)'] == 100]
sns.countplot(x = 'Gender' , data = df )
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.barplot(x='Profession', y='Age',hue = 'Gender', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(data=df, x='Age',color='purple')
```

```python
sns.boxplot(data=df, x='Profession', y='Annual Income ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.violinplot(data=df, x='Profession', y='Annual Income ($)')
plt.xticks(rotation=90)
plt.title('VIOLIN PLOT SHOWING ANNUAL INCOME OF PROFESSIONS')
plt.show()
```

```python
sns.pointplot(x='Profession', y ='Spending Score (1-100)', data =df, hue='Gender', ci=None)
plt.title('MEAN SPENDING SCORES OF PROFESSIONS')
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

```python
sns.scatterplot(data = df, x = 'Annual Income ($)', y = 'Spending Score (1-100)')
plt.title('SCATTER PLOT SHOWING RELATIONSHIP BETWEEN ANNUAL INCOME AND SPENDING SCORE')
plt.show()
```

```python
eng = df[df['Profession'].isin(['Engineer'])]
eng.head(10)
sns.lineplot(data=eng, x ='Annual Income ($)', y ='Spending Score (1-100)',hue= 'Gender')
```

```python
h_score = df[df['Spending Score (1-100)'] == 100]
print(h_score)
h_score['Profession'].value_counts(sort = 'ascending').plot.bar()
```

```python
mask = np.triu(np.ones_like(df.corr(),dtype = bool))
heatmap = sns.heatmap(df.corr(),vmin =-1,mask=mask, vmax=1, annot =True, cmap='Blues')
heatmap.set_title('Triangle Correlation Heatmap',fontdict={'fontsize':14},pad=18)
```
